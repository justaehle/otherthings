```python
#Moebiusband
%matplotlib notebook
from pylab import *
from mpl_toolkits.mplot3d import Axes3D

Lp = 200
Lq = 200
p = np.linspace(0,pi,100)
q = np.linspace(0,2*pi,100)
r = 1
P,Q = np.meshgrid(p,q) # (P[i,j],Q[i,j]) = (p[i],q[j])

x = cos(P)*(1+r/2*cos(P/2)) # x[i,j] = x(p[i],q[j])
y = sin(P)*(1+r/2*cos(P/2))
z = r/2*sin(P/2)

fig=figure()
ax3d=fig.add_subplot(111,projection='3d',aspect=1)
ax3d.plot_surface(x,y,z,rstride=10,cstride=10,alpha=0.5,color='red')    # oder plot_wireframe(...)


ax3d.set_xlim(-5,5) # man sollte moeglichst gleich grosse Koordinatenbereiche waehlen,
ax3d.set_ylim(-5,5) # um Verzerrungen bei der Darstellung zu vermeiden
ax3d.set_zlim(-5,5)

#plotaufbereitung

show()
```
